```python
import csv
import math
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
from sklearn.cluster import KMeans


def load_data(new_file):
    with open(new_file,'r')as file:
        read_file = csv.reader(file)
        data = []

        for ele in read_file:
            data.append(ele)
 
    train_data = []
    for item in data:
        train_data.append(np.array([float(i) for i in item]))
    train_data = np.vstack(train_data)

    feature_train = train_data[:,:-1]
    label_train = train_data[:,-1]
    return feature_train, label_train, train_data


def rbf_layer1(X, mu, gamma):
    """Layer 1: Get transformed phi(x)"""
    phi_x_ = np.zeros((len(X), len(mu)))
    for i in range(len(X)):
        for j in range(len(mu)):
            xmu = np.matmul(np.transpose(X[i] - mu[j]),(X[i] - mu[j]))
            phi_x_[i][j] = math.exp(-gamma*(xmu))
    return phi_x_



def cal_MSE(_Y, Y):
    l = len(Y)
    res = np.matmul(np.transpose(Y-_Y),(Y-_Y))
    return res/l


def calculate_gamma(X, M):
    """Calculate gamma"""
    
    avg_spacing = cal_avg_spacing(X, M)
    gamma_ = 1/(2*((2*avg_spacing)**2))
    return gamma_


def cal_avg_spacing(X, M):
    """Calculate average spacing"""
    widthSpace = 1
    for i in range(len(X[0])):
        cur_width = (max(X[:][i]) - min(X[:][i]))**(1/len(X[0]))
       
        widthSpace = widthSpace*cur_width
    
    return (widthSpace)## / (M**(1/len(X[0])))


def Polynomial(X,degree):
    model = PolynomialFeatures(degree)
    update_X = model.fit_transform(X)
    return update_X
```

```python
def select_feature(X):
    first_two = []
    for item in X:
        first_two.append(item[:-1])
    return np.vstack(first_two)

def select_label(data):
    label = []
    for item in data:
        label.append(item[-1])
    return np.vstack(label)

def make_subset(X,n):
    k, m = divmod(len(X),n)
    return list(X[i*k+min(i,m) : (i+1)*k+min(i+1,m)] for i in range(n))

def combine_data(data,c):
    new_data = []
    for i in range(len(data)):
        if i!=c:
            new_data.append(data[i])
    return np.concatenate(new_data,axis=0)
        
def unbiased_std(x):
    avg = np.mean(x);   std = 0
    for i in range(len(x)):
        std += (x[i]-avg)*(x[i]-avg)
    return avg, np.sqrt(std/(len(x)-1))

def cross_val(k, data):
    data = np.random.permutation(data)
    data_set = make_subset(data,k)
    mae_set = []; r2_set = []; pmse_set = []; pmae_set = []; mr2_set = []
    for i in range(k):
        #process data
        data_val = data_set[i]
        data_tr = combine_data(data_set,i)
        X_val = select_feature(data_val);  Y_val = select_label(data_val)
        X_tr = select_feature(data_tr);  Y_tr = select_label(data_tr)
        
        #make model and predict
        reg = LinearRegression().fit(X_tr, Y_tr)
        Y_val_pred = np.vstack(reg.predict(X_val))
        
        mae_set.append(mean_absolute_error(Y_val_pred, Y_val))
        r2_set.append(r_squared(Y_val_pred, Y_val))
        pmse_set.append(pMSE(Y_val_pred, Y_val, r=10))
        pmae_set.append(pMAE(Y_val_pred, Y_val, r=10))
        mr2_set.append(m_r_squared(Y_val_pred, Y_val, r=10))
        
    return np.mean(mae_set), np.mean(r2_set), np.mean(pmse_set), np.mean(pmae_set), np.mean(mr2_set)

def find_gamma(tr_X_aug, mu, tr_Y, gamma):
    gamma_set = [gamma/4, gamma/2, gamma, gamma*2, gamma*4]
    mae_set = np.zeros(5); r2_set = np.zeros(5); pmse_set = np.zeros(5); pmae_set = np.zeros(5); mr2_set = np.zeros(5)

    for i in range(len(gamma_set)):
        phi_X = rbf_layer1(tr_X_aug, mu, gamma_set[i])
        phi_X = Polynomial(phi_X,1)
        phi_data = np.concatenate((phi_X, tr_Y),axis=1)
        
        mae, r2, pmse, pmae, mr2 = cross_val(5, phi_data)
        mae_set[i] = mae; r2_set[i] = r2; pmse_set[i] = pmse; pmae_set[i] = pmae; mr2_set[i] = mr2
    return mae_set, r2_set, pmse_set, pmae_set, mr2_set, gamma_set, np.mean(mae_set)

def rand_select_train(tr_X, tr_Y, c):
    tr_X_aug = Polynomial(tr_X,1)
    mu, gamma = get_mu_gamma(tr_X, tr_X_aug, c)
    mae_set, r2_set, pmse_set, pmae_set, mr2_set, gamma_set, mean_mae = find_gamma(tr_X_aug, mu, tr_Y, gamma)
    return np.mean(mae_set), np.mean(r2_set)#, np.mean(pmse_set), np.mean(pmae_set), np.mean(mr2_set)

def kmean_select_train(tr_X, tr_Y, c):
    gamma = calculate_gamma(tr_X, c)
    kmeans = KMeans(n_clusters=c, init='random', random_state=0).fit(tr_X)
    center_ = Polynomial(kmeans.cluster_centers_,1)
    tr_X_aug = Polynomial(tr_X,1)
    mae_set, r2_set, pmse_set, pmae_set, mr2_set, gamma_set, mean_mae = find_gamma(tr_X_aug, center_, tr_Y, gamma)
    return mean_mae, np.mean(pmse_set)

def get_mu_gamma(tr_X, tr_X_aug, c):
    rand_idx = np.random.choice(len(tr_X), c, replace=False)
    gamma = calculate_gamma(tr_X, c)
    mu = []
    for i in rand_idx:
        mu.append(tr_X_aug[i])
    mu = np.vstack(mu)
    return mu, gamma

def get_test_MSE(tr_X_aug, tr_Y, te_X_aug, te_Y, mu, gamma):
    tr_phi_X = Polynomial(rbf_layer1(tr_X_aug, mu, gamma), 1)
    reg = LinearRegression().fit(tr_phi_X, tr_Y)
    
    te_phi_X = Polynomial(rbf_layer1(te_X_aug, mu, gamma), 1)
    te_Y_pred = np.vstack(reg.predict(te_phi_X))
    return cal_MSE(te_Y_pred, te_Y)
  
```
